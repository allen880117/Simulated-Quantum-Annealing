```python
import numpy as np
import math
import matplotlib.pyplot as plt
import statistics as st
```

```python
energy_list = []
best_list = []
with open("out.txt") as file:
    for line in file:
        if (line.find("SUM:")!=-1):
            tmp = float(line.split()[1])
            energy_list.append(tmp)
        if (line.find("BEST:")!=-1):
            tmp = float(line.split()[1])
            best_list.append(tmp)
```

```python
plt.figure(figsize=(20,10))
plt.plot(energy_list[:], 'o-')
plt.plot(best_list[:], 'o-')
```
